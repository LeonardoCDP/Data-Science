# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

#Obtenção dos Dados

Devido a dificuldade de encontrar um base de dados no Brasil completa que não tenha um conflito de informações entre o governo federal ,municipal e estadual nas informações, por ser uma doença nova com um baixo ou nem um número de testes e novos testes, iremos utilizar a base de dados mundial da Our World in Data.

* Para acessar o site da Our World in Data [Clique Aqui](https://ourworldindata.org/)

* Para acessar o dicionário original das colunas [Clique Aqui](https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data-codebook.md)

#Dicionário das Colunas

* iso_code - ISO 3166 Código do País
* continent - Continente
* location - Localização Geográfica(País)
* date - Data do Registro
* total_cases - Total de Casos Confirmados
* new_cases - Novos casos Confirmados
* total_deaths - Total de Mortes Confirmadas
* new_deaths - Novas Mortes Confirmadas
* total_cases_per_million - Total de Casos Confirmados Por Um Milhão de Pessoas
* new_cases_per_million - Novos Casos Confirmados Por Um Milhão de Pessoas
* total_deaths_per_million - Total de Mortes Confirmadas Por Um Milhão de Pessoas
* new_deaths_per_million - Novas Mortes Confirmadas Por Um Milhão de Pessoas
* new_tests - Novos testes
* total_tests - Total de Testes
* total_tests_per_thousand - Total de Testes Por Um Mil de Pessoas
* new_tests_per_thousand - Novos Teste Por Um Mil de Pessoas
* new_tests_smoothed - Novos Testes Acumulados Por Sete Dias Consecutivos
* new_tests_smoothed_per_thousand - Novos Testes Acumulados Por Sete Dias Consecutivos Por Um Mil de Pessoas
* tests_units - Unidades Utilizadas Por Região Para Relatar Seus Testes
* stringency_index - Índice Informado Pelo Governo Baseado Em 9 Indicadores de Isolamento Social
* population - Tamanho da População em 2020
* population_density - Índice de Densidade Populacional Em 2020
* median_age - Idade Média
* aged_65_older - Proporção da População Com 60 Anos ou Mais Registrados Em 2017
* aged_70_older - Proporção da População Com 70 Anos ou Mais Registrados Em 2017
* gdp_per_capita - Produto Interno Bruto Do País Medido Em Dólar(PIB Registrado em 2011 Último Ano Disponível)
* extreme_poverty - Índice de Pessoas Que Vivem Em Extrema Pobreza Registrado em 2010
* cardiovasc_death_rate - Taxa de Mortalidade de Mortes Causadas Por Doenças Cardíacas em 2017(Numero Anual de Mortes Por Cem Mil pessoas)
* diabetes_prevalence - Proporção de Diabéticos da População de 20 a 79 Anos de Idade
* female_smokers - Porcentagem de Mulheres Fumantes(Último Ano Registrado)
* male_smokers - Porcentagem de Homens Fumantes(Último Ano Registrado)
* handwashing_facilities - Porcentagem de Pessoas Com Higiene Básica Disponível
* hospital_beds_per_thousand - Número de Camas Em Hospitais Por Um Mil de Pessoas(Último Ano Disponível 2010)
* life_expectancy - Expectativa de Vidas Dos Nascidos Em 2019

```
# importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importando o dataset COVID-19 e criando um Data Frame
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

Obs: Alguns países possuem um índice de novos teste informados que torna a análise exploratória deles mais conclusiva fora do âmbito mundial, no caso iremos a análise exploratória do Brasil.

```
# visualizando os 5 países que possuem o maior número de casos registrados no último período
df.loc[df.date == '2020-08-04'].sort_values(by='total_cases', ascending=False)[1:6]
```

Os Estados Unidos possuem o maior número de casos seguido do Brasil, Índia,Rússia, África do Sul, etc.

```
# visualizando os 5 países com o maior número de mortes registradas
df.loc[df.date == '2020-08-04'].sort_values(by='total_deaths', ascending=False)[1:6]
```

Os Estados unidos possui o maior numero de mortes registradas seguido por Brasil, Mexico, Reino Unido, Índia,etc.

```
# gerando um Data Frame com os top 5 países com mais mortes registradas
top5_total_deaths = df.loc[df.date == '2020-08-04', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[1:6].copy()

# plotando um gráfico de barras com os top 5 paises com maior número de mortes registradas
fig, ax = plt.subplots(figsize=(10, 5))

sns.barplot(x='location', y='total_deaths', data=top5_total_deaths, ax=ax)
ax.set_ylabel('Total de Mortes')
ax.set_xlabel('Paises')
ax.set_title('Top 5 Paises com Maior Número de Mortes Causadas Pelo COVID - 19')

plt.show()
```

Podemos ver que os Estados Unidos e o país com o maior número de mortes causadas pelo COVID - 19 no mundo.

```
# visualizando os top 5 países com o maior número de mortos por milhão de pessoas
df.loc[df.date == '2020-08-04'].sort_values(by='total_deaths_per_million', ascending=False)[1:6]
```

Podemos ver que Bélgica e o país com o maior número de mortes por milhão de pessoas seguido do Reino Unido, Andorra, Peru e Ítalia.

```
# gerando um Dada Frame com os top 5 Países com maior número mortos por milhão de pessoas
top_5_mortes_por_milhao = df.loc[df.date == '2020-08-04', ['location', 'total_deaths_per_million']].sort_values(by='total_deaths_per_million', ascending=False)[1:6].copy()

# plotando um gráfico de barras para os Top 5 Países Com o Maior Número de Mortos Por milhão de Pessoas
fig, ax = plt.subplots(figsize=(10,5))

sns.barplot(x='location', y='total_deaths_per_million', data=top_5_mortes_por_milhao, ax=ax)
ax.set_xlabel('Países')
ax.set_ylabel('Mortos Por Milhão de Pessoas')
ax.set_title('Top 5 Países Com Maior Número de Mortos Por Milhão de Pessoas')
plt.show()
```

Podemos ver que quando analisamos o número de mortos por milhão de pessoas os países com mais mortos mudam, trazendo a Bélgica para o primeiro lugar da lista.

```
# gerando um Data Frame para analisar os dados mundiais
df_world = df.loc[df.location == 'World'].copy()

# plotando gráfico de linhas para visualizar a evolução do casos de covid no mundo
fig, ax = plt.subplots(figsize=(10,5))

sns.lineplot(x='date' , y='total_cases', data=df_world, ax=ax)
ax.set_ylabel('Total de Casos')
ax.set_xlabel('Data')
ax.set_title('Evolução dos Casos de COVID - 19 no Mundo')
plt.show()
```

Podemos notar que a partir do mês de Março de 2020 os casos de COVID - 19 começaram a aumentar seus registros no mundo.

```
# pegando os dados do último dia registrado no Mundo
df_world_ultimo_dia = df_world.loc[df_world.date == '2020-08-04'].copy()
# verificando a porcentagem de mortes X casos no Mundo
mortes = df_world_ultimo_dia.total_deaths.reset_index(drop=True)
casos = df_world_ultimo_dia.total_cases.reset_index(drop=True)
registros = df_world.shape[0]

print(f'{casos[0]:.0f} Casos registrados no Mundo')
print(f'{mortes[0]:.0f} Mortes registradas no Mundo')
print(f'Média de {mortes[0]/registros:.2f} mortes diarias no Mundo')
print(f'Média de {casos[0]/registros:.2f} casos diarios no Mundo')
print(f'O numero de mortes registradas representa {(mortes[0]*100)/casos[0]:.2f}% do numero de casos')
```

Foram registrados um total de 18263542 casos, 693726 mortes causados pelo COVID - 19 com uma média de 83777.72 casos e 3182.23 mortes diariamente no Mundo.

A porcentagem de mortes e de 3.80% do total de casos registrados pelo COVID - 19 no mundo.

```
# gerando um Data Frame para análise do PIB dos Países X Total de Mortes e Casos
df_per_capita = df.loc[df.date == '2020-08-04'].sort_values(by='total_deaths', ascending=False)[1:].copy()

# plotando um gráfico de dispersão para comparação de Total de Mortes X PIB do País.
fig, ax = plt.subplots(figsize=(10,5))

sns.scatterplot(x='total_deaths', y='gdp_per_capita', data=df_per_capita, ax=ax)
ax.set_ylabel('PIB do País')
ax.set_xlabel('Total de Mortes')
ax.set_title('Total de Mortes X PIB do País')
plt.show()
```

Analisando o gráfico de dispersão notamos que a relação do PIB do País em relação a quantidade de mortes causadas pelo COVID - 19 não é conclusiva, pois envolve fatores como proporção de testes,etc.

```
# plotando um gráfico de dispersão para comparação do Total de Casos X PIB do País.
fig, ax = plt.subplots(figsize=(10,5))

sns.scatterplot(x='total_cases', y='gdp_per_capita', data=df_per_capita, ax=ax)
ax.set_ylabel('PIB do País')
ax.set_xlabel('Total de Casos')
ax.set_title('Total de Casos X PIB do País')
plt.show()
```

Analisando o gráfico de dispersão notamos que a relação do PIB do País em relação a quantidade de infectadas pelo COVID - 19 não é conclusiva, pois envolve fatores como proporção de testes,etc.

```
# gerando um Data Frame para análise da correlação entre casos por milhão de pessoas vs densidade populacional
df_populacao = df.loc[df.date == '2020-08-04', ['total_cases_per_million','population_density']].sort_values(by='population_density', ascending=False)[1:].copy()

# plotando um gráfico de dispersão para analisar a correlação de casos por milhão de pessoas vs densidade populacional
fig, ax = plt.subplots(figsize=(10,5))

sns.scatterplot(x='population_density', y='total_cases_per_million', data=df_populacao, ax=ax)
ax.set_xlabel('Densidade Populacional')
ax.set_ylabel('Total de Casos Por Milhão de Pessoas')
ax.set_title('Total de Casos Por Milhão de Pessoas X Densidade Populacional')
plt.show()
```

Analisando o gráfico de dispersão notamos que a relação da densidade populacional em relação a quantidade de casos por milhão de pessoas infectadas pelo COVID - 19 não é conclusiva, pois envolve fatores como proporção de testes,etc.

### Análise Exploratória para o Brasil

```
# visualizando a quantidade de dados ausentes ou nulos registrados no Data Frame do Brasil
(df_brasil.isnull().sum()/ df_brasil.shape[0]).sort_values(ascending=False)
```

Podemos ver que o Brasil possui 100% de dados nulos ou ausente na coluna de novos testes realizados e novos testes realizados por um mil de pessoas, na coluna total de teste e total de testes por um mil pessoas um total de 97% de dados nulos ou ausentes, na coluna novos teste acumulados em sete dias e novos testes acumulados em sete dias por um mil de pessoas 89% dos dados nulos ou ausentes.

A primeira morte de causada pelo COVID-19 foi registrada no dia 18 de Março de 2020 a qual veio a ocorrer 21 dias após o primeiro caso registrado no dia 26 de Fevereiro de 2020.

```
# plotando um gráfico de linhas para analisar o Total de Casos X Total de Mortes
fig, ax = plt.subplots(figsize=(10,5))

sns.lineplot(x='date', y='total_cases', data=df_brasil, ax=ax)
sns.lineplot(x='date', y='total_deaths', data=df_brasil, ax=ax)
ax.legend(('Total de Casos', 'Total de Mortes'))
ax.set_title('Total de Casos X Total de Mortes')
ax.set_ylabel('Quantidade de Pessoas')
ax.set_xlabel('Data')

plt.show()
```

Analisando o gráfico de linhas vemos que a quantidade de casos do COVID - 19 no Brasil vem sofrendo uma evolução considerável assim como a quantidade de mortes vem crescendo lentamente em comparação com a quantidade de casos.

```
# pegando os dados do ultimo dia registrado no Brasil
df_brasil_ultimo_dia = df_brasil.loc[df_brasil.date == '2020-08-04'].copy()
# verificando a porcentagem de Brasil x casos no Brasil
mortes = df_brasil_ultimo_dia.total_deaths.reset_index(drop=True)
casos = df_brasil_ultimo_dia.total_cases.reset_index(drop=True)
registros = df_brasil.shape[0]

print(f'{casos[0]:.0f} Casos registrados no Brasil')
print(f'{mortes[0]:.0f} Mortes registradas no Brasil')
print(f'Média de {mortes[0]/registros:.2f} mortes diarias no Brasil')
print(f'Média de {casos[0]/registros:.2f} casos diarios no Brasil')
print(f'O numero de mortes registradas representa {(mortes[0]*100)/casos[0]:.2f}% do numero de casos')
```

O número de casos registrados e de 2750318 casos e 94665 mortes totais causadas pelo COVID - 19 uma média de 434.24 mortes e 12616.14 casos registrados diariamente.

O percentual de mortes registradas pelo COVID - 19 representa 3.44% do total de casos registrados.

## Conclusão

Podemos concluir que o COVID - 19 por ser um vírus com alto índice de transmissão,contágio e por não ter uma vacina torna-se de difícil controle de propagação mesmo utilizando o métodos como: quarentena , controle de aglomerações , uso de equipamento de controle de transmissão(máscaras, luvas, álcool, limpeza dos ambientes, sabão, etc...), os quais não se mostram 100% eficientes para controle da propagação.

Por ser um vírus que possui sintomas diferentes em cada paciente infectado que em alguns casos pode ser assintomático e não vir a apresentar sintomas ou apresentar sintomas leves e ter um alto custo de teste variando de local para local, o'que não garante que um paciente testado negativo hoje , não possa se tornar positivo no dia seguinte,devido a ele não possuir resistência imunológica contra o vírus devido a falta de uma vacina para o mesmo.

Os tratamentos utilizados mostraram-se eficientes ocasionando uma leve redução no número de novas de mortes.

O percentual de mortes no Mundo e de 3.80% dos casos totais registrados.

O percentual de mortes no Brasil e de 3.44% dos casos totais registrados.

O Brasil teve o seu primeiro caso de COVID - 19 registrado no dia 26 de fevereiro de 2020 vindo a ter o seu primeiro óbito 21 dias depois no dia 18 de Março de 2020.

A primeira morte de causada pelo COVID-19 foi registrada no dia 18 de Março de 2020 a qual veio a ocorrer 21 dias após o primeiro caso registrado no dia 26 de Fevereiro de 2020.

Levando em conta que o tratamento previne a morte do paciente e não a transmissão e o contágio.
